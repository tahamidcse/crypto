# Transposition Cipher
Suppose you are given a line of text as a plaintext, find out the corresponding Transposition Cipher. Take width as input. Then perform the reverse operation to get original plaintext.

```python
# Encryption

def encryption(plain_text, width = 4):
    length = len(plain_text)
    cipher_text = ""

    for k in range(width):
        for i in range(k, length, width):
            cipher_text += plain_text[i]
    
    return cipher_text
```

```python
# Decryption

def decryption(cipher_text, width = 4):
    length = len(cipher_text)
    plain_text = [''] * length
    idx = 0

    for k in range(width):
        for i in range(k , length, width):
            plain_text[i] = cipher_text[idx]
            idx += 1

    return ''.join(plain_text)
```

```python
# 3. Single Columnar Transposition
plain_text = "Computer Science and Engineering, University of Rajshahi, Bangladesh."
width = 3
print("Original Text : ", plain_text)

cipher_text = encryption(plain_text)
print("Cipher Text : " , cipher_text)

decrypted_msg = decryption(cipher_text)
print("Decrypted msg : ", decrypted_msg)
```

# Double Transposition

```python
# Function for Double Transposition Encryption

def double_transposition_encryption(plain_text, width_1 = 4 , width_2 = 3):
    first_cipher = encryption(plain_text,width_1)
    second_cipher = encryption(first_cipher, width_2)
    return first_cipher, second_cipher
```

```python
# Function for Double Transpositon Decryption:

def double_transposition_decryption(cipher_text,width_1 = 4, width_2 = 3):
    first_decrypt_cipher = decryption(cipher_text,width_2)
    plain_text = decryption(first_decrypt_cipher,width_1)
    return first_decrypt_cipher, plain_text
```

```python
# main dtc
plain_text = "Computer Science and Engineering, University of Rajshahi, Bangladesh."

print("Original : ",plain_text)

encrypt_1, encrypt_2 = double_transposition_encryption(plain_text)
print("First Encryption : ", encrypt_1, "\nSecond Encryption : ", encrypt_2)

decrypt_1, d_2 = double_transposition_decryption(encrypt_2)
print("First Decrypion : ", decrypt_1, "\nSecond Decryption : ", d_2)
```
