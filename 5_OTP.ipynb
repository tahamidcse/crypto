# 5. One Time Pad
You are supplied a file of large non-repeating set of truly random key letter. Your job is to encrypt the plaintext using One Time Pad technique. Then perform the reverse operation to get original plaintext.

```python
key = ""
store_key = ""

with open("OTP","r") as f:
    key = f.read()
    print(key)
store_key = key
```

```python
# Encryption
def encryption(plain_text, key):
    plain_text = plain_text.upper()   # normalize
    # key = key.upper()
    cipher_text = ""
    used_key = ""
    for i, ch in enumerate(plain_text):
        char = (ord(ch) - ord('A') + (ord(key[i]) - ord('A'))) % 26
        cipher_text += chr(ord('A') + char)
        used_key += key[i]
    return cipher_text, used_key
```

```python
# Decryption
def decryption(cipher_text, key):
    cipher_text = cipher_text.upper()  # normalize
    # key = key.upper()
    plain_text = ""
    for i, ch in enumerate(cipher_text):
        char = (ord(ch) - ord('A') - (ord(key[i]) - ord('A'))) % 26
        plain_text += chr(ord('A') + char)
    return plain_text
```

```python
# main
plain_text = "bangladesh"
cipher_text,ukey = encryption(plain_text, key)
decrypted_text = decryption(cipher_text,key)

remaining_key = key[len(plain_text):]

with open("OTP", "w") as f:
    f.write(remaining_key)

print("Original :" , plain_text)
print("Cipher Text :" , cipher_text)
print("Decrypted message :" ,decrypted_text)
print("Key : " , ukey)
print("Original Key :", key)
print("Remaining key : ", remaining_key[:len(key)],"...")
```
