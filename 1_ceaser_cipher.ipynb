# Ceaser Cipher
Suppose you are given a line of text as a plaintext, find out the corresponding Caesar Cipher. Then perform the reverse operation to get original plaintext.

```python
# Function for Encryption
def encryption(plain_text):
    cipher_text = ""
    for ch in plain_text:
        if ch.isalpha():
            ch = chr(ord(ch) + 3)
            if ch.isalpha() != True:
                ch = chr(ord(ch) - 26)
        cipher_text += ch
    
    return cipher_text
```

```python
# Function for Decryption
def decryption(cipher_text):
    plain_text = ""

    for ch in cipher_text:
        if ch.isalpha():
            ch = chr(ord(ch) - 3)
            if ch.isalpha() != True:
                ch = chr(ord(ch) + 26)
        plain_text += ch
    # print(plain_text)
    return plain_text
```

```python
# Main Function
original_text = "Computer Science and Engineering, University of Rajshahi"
print("original : ", original_text)

encrypt = encryption(original_text)
print("Encypted Text : " , encrypt)

decrypt = decryption(encrypt)
print("Decrypted Message : ", decrypt)

if original_text == decrypt:
    print("Your ceaser cipher is working well")
else:
    print("Decrypted msg does not match with the original msg")
```
