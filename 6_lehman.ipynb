# Lehmann Primality Test

```python
import random

def lehmann_primality_test(p, t):
    if p < 2:
        return False
    if p in [2, 3]:
        return True
    if p % 2 == 0:
        return False  # even numbers > 2 are composite

    for _ in range(t):
        a = random.randint(2, p - 2)
        x = pow(a, (p - 1) // 2, p)  # a^((p-1)/2) mod p
        if x not in [1, p - 1]:
            return False  # definitely composite
    return True  # probably prime
```

```python
p = 15
res = lehmann_primality_test(p,t=10)
# print(res)
print(f"{p} is probably prime" if res else f"{p} is composite")
```
