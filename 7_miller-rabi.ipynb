# miller_rabin

```python
import random

def miller_rabin(n,k=5):
    if n < 2:
        return False
    if n in [2,3]:
        return True
    if n%2 == 0:
        return False
    
    # step 1: write n-1 as (2^b) * m with m odd
    m = n-1
    b = 0
    while m % 2 == 0: # count how many times 2 divides n-1
        m //= 2
        b += 1

    # perform k independent test for accuracy
    for _ in range(k):
        a = random.randrange(2,n-1) # step 2 : choose random base a in [2,n-2]

        z = pow(a,m,n) # step 3 : z = a^m mod n

        if z == 1 or z == n-1: # step 4 : if z == 1 or z == n-1, test passes this round
            continue

        # step 5: repeat squaring up to b-1 times
        for _ in range(b-1):
            z = (z * z) % n
            if z == n-1:
                break
            if z == 1:
                return False
            
        else:
            return False # if loop ended without finding n-1, composite
        
    return True
```

```python
# main
test_numbers = [7,13,15,17,19,21,561]
for num in test_numbers:
    print(f"{num} is probably prime " if miller_rabin(num, k =5) else f"{num} is composite" )
```
