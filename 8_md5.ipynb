```python
import hashlib

def md5_hash(data: bytes) -> str:
    """Return the MD5 hex digest of data (bytes)."""
    m = hashlib.md5()
    m.update(data)
    return m.hexdigest()

# Example usage
if __name__ == "__main__":
    text = "hello world"
    digest = md5_hash(text.encode('utf-8'))
    print("MD5(\"{}\") = {}".format(text, digest))
```
