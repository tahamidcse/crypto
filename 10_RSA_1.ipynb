# 10. RSA

```python
import math
import random

def gcd(a,b):
    while b:
        a , b = b, a%b
    
    return a

def mod_inverse(e,phi):
    def egcd(a,b):
        if a == 0:
            return b , 0 ,1
        gcd, x1, y1 = egcd(b%a,a)
        x = y1 - (b//a) * x1
        y = x1
        return gcd, x, y
    gcd, x , y = egcd(e,phi)
    return (x%phi+phi) % phi

def is_prime(n):
    if n < 2:
        return False
    for i in range(2,int(n**0.5) + 1):
        if n%i == 0:
            return False
    return True
```

```python
# generate key pair
def generate_key_pair():
    primes = [x for x in range(50, 200) if is_prime(x)]
    p = random.choice(primes)
    q = random.choice(primes)
    while p == q:
        q = random.choice(primes)

    n = p * q
    phi = (p - 1) * (q - 1)

    e = 65537
    if gcd(e, phi) != 1:
        e = random.randint(3, phi-1)
        while gcd(e, phi) != 1:
            e = random.randint(3, phi-1)

    d = mod_inverse(e, phi)
    return (e, n), (d, n), p, q


def rsa_encrypt(message, public_key):
    e,n = public_key
    return pow(message,e,n)

def rsa_decrypt(cipher_text, private_key):
    d,n = private_key
    return pow(cipher_text,d,n)

def string_to_ascii(text):
    return [ord(char) for char in text]

def ascii_to_string(ascii_list):
    return ''.join(chr(num) for num in ascii_list)

pu_k,pr_k,p,q = generate_key_pair()
e,n = pu_k
d,n = pr_k

print("=====RSA key generation=====")
print("p = ", p)
print("q = ", q)
print("n = p * q : ",n)
print(f"Public Key (e,n):({e},{n})")
print(f"private Key (d,n) = ({d},{n})")
print()
```

```python
# main
print("-----Integer Encryption and Decryption-----")
original_message = 42
print(f"Original message : {original_message}")

encrypted_message = rsa_encrypt(original_message, public_key = pu_k)
print(f"Encrypted message : {encrypted_message}")

decrypted_message = rsa_decrypt(encrypted_message,private_key=pr_k)
print(f"Decrypted message : {decrypted_message}")


print("-----String Encryption and Decryption")
string_message = "rescue"
print(f"original_message: {string_message}")

string_ascii = string_to_ascii(string_message)
print(f"Original ascii : {string_ascii}")
encrypted_chars = []
for ascii_val in string_ascii:
    if ascii_val < n:
        encrypted_char = rsa_encrypt(ascii_val,public_key=pu_k)
        encrypted_chars.append(encrypted_char)
    else:
        print(f"Something went wrong")

print(f"Encrypted ascii : {encrypted_chars}")

decrypted_ascii = []
for ds in encrypted_chars:
    dc = rsa_decrypt(ds,private_key=pr_k)
    decrypted_ascii.append(dc)

print(f"Decrypted ascii : {decrypted_ascii}")
decrypted_string = ascii_to_string(decrypted_ascii)
print(f"Decrypted message : {decrypted_string}")
```
